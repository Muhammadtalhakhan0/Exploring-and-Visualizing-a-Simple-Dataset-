# Exploring and Visualizing a Simple Dataset
**1: Objective** The objective of this task is to learn how to load, inspect, and visualize a dataset in order to understand data trends and distributions.

**2: Dataset** The Iris dataset is used in this task.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**3: Loading the Dataset**

The Iris dataset is loaded using the seaborn library.

```python
iris = sns.load_dataset("iris")
```

**4: Inspecting the Dataset**

 We check the shape, column names, and first few rows of the dataset.

```python
# Shape of the dataset
iris.shape
```

```python
# Column names
iris.columns
```

```python
# First five rows
iris.head()
```

*Explanation of **iris.head()** Output*

The iris.head() command displays the first five rows of the Iris dataset. Each row represents one flower sample, and each column shows a specific feature: sepal length, sepal width, petal length, petal width, and the flower species. From this output, we can see that the first five samples belong to the setosa species and we can understand the basic structure and values of the dataset.

**5: Dataset Information**

The .info() method shows dataset structure.

```python
iris.info()
```

*Explanation of **iris.info()** Output*

The iris.info() command shows a summary of the dataset. It tells us that the dataset has 150 rows and 5 columns. All columns have 150 non-null values, which means there are no missing values in the dataset. Four columns are of type float (numerical features) and one column is of type object (the species name). It also shows that the dataset uses about 6 KB of memory.

**6: Summary Statistics**

The describe() function generates statistical information such as mean,
minimum, maximum, and standard deviation of numerical features.


```python
iris.describe()
```

*Explanation of **iris.describe()** Output*

The iris.describe() command gives a statistical summary of the numerical columns in the dataset. It shows that each feature has 150 values. The table displays the average (mean), spread (standard deviation), minimum and maximum values, and the 25%, 50%, and 75% percentiles. This helps us understand the range and distribution of sepal and petal measurements in the dataset.

**7: Scatter Plot**

A scatter plot is used to visualize the relationship between petal length and petal width.

```python
sns.scatterplot(
    data=iris,
    x="petal_length",
    y="petal_width",
    hue="species"
)

plt.title("Scatter Plot of Petal Length vs Petal Width")
plt.show()
```

*Explanation of **Scatter Plot** (Petal Length vs Petal Width)*

This scatter plot shows the relationship between petal length and petal width for three Iris species. Each dot represents one flower, and different colors show different species (setosa, versicolor, virginica). We can see that setosa forms a separate cluster with small petals, while versicolor and virginica have larger petals and form different groups, which means these features help in distinguishing the species.

**8: Histograms**

Histograms are used to show the distribution of feature values.

```python
iris.hist(figsize=(10, 8))
plt.show()
```

*Explanation of **Histograms** Output*

This output shows four histograms, one for each feature: sepal length, sepal width, petal length, and petal width. Each histogram displays how the values are distributed and how frequently they occur. From these graphs, we can see the range of values and which measurements are more common, helping us understand the spread and distribution of the data.

**9: Box Plot**

Box plots are used to identify outliers and understand data spread.


```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris)
plt.show()
```

*Explanation of **Box Plot** Output*

This box plot shows the distribution of four features: sepal length, sepal width, petal length, and petal width. Each box represents the spread of the data, with the line inside showing the median value. The dots outside the boxes indicate outliers. From this plot, we can see how the values vary for each feature and where some unusual values exist.
